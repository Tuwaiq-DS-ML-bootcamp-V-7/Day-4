### Task 1
1. Write a function named is_prime that takes a number as a parameter and returns True if the number is prime and False otherwise. Print whether the number 29 is prime.

```python
def is_prime(number):
    if number <= 1:
        return False
    if number == 2:
        return True  # 2 is the only even prime number
    if number % 2 == 0:
        return False  # Even numbers greater than 2 are not prime

    # Check for odd factors from 3 up to the square root of the number
    for i in range(3, int(number**0.5) + 1, 2):
        if number % i == 0:
            return False

    return True

# Check if 29 is prime
number_to_check = 29
if is_prime(number_to_check):
    print(f"{number_to_check} is a prime number.")
else:
    print(f"{number_to_check} is not a prime number.")
```

### Task 2
1. Create a list of 10 numbers and sort them in descending order using the sort() function.
2. Use the len() function to print the length of the sorted list.
3. Use the zip() function to combine three lists (names, ages, and cities) into a list of tuples.
4. Use the range() function to create a list of even numbers from 2 to 20 and print the list.

```python
# Create a list of 10 numbers
numbers = [23, 45, 12, 67, 5, 89, 34, 56, 78, 9]

# Sort and calculate the length of the list in descending order
numbers.sort(reverse=True)
sorted_length = len(numbers)

# Print the sorted list
print("Sorted numbers in descending order:", numbers)
print("Length of the sorted list:", sorted_length)
```

```python
# Sample lists for names, ages, and cities
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 28]
cities = ["New York", "Los Angeles", "Chicago"]

# Use zip() to combine the lists into a list of tuples
combined_list = list(zip(names, ages, cities))

# Print the list of tuples
print("Combined list of tuples:")
for item in combined_list:
    print(item)
```

```python
# Create a list of even numbers from 2 to 20 using range()
even_numbers = list(range(2, 21, 2))

# Print the list of even numbers
print("List of even numbers from 2 to 20:", even_numbers)
```

### Task 3
Build a phone book program that receives the phone number, and returns the name of the owner.
You can follow the table below:

| Name     | Number     |
| -------- | ---------- |
| Amal     | 1111111111 |
| Mohammed | 2222222222 |
| Khadijah | 3333333333 |
| Abdullah | 4444444444 |
| Rawan    | 5555555555 |
| Faisal   | 6666666666 |
| Layla    | 7777777777 |

- If the number exists, print the owner. Otherwise, print "Sorry, the number is not found".
- If the number is less or more than 10 numbers, print "This is invalid number".
- If the number contains letters or symbols, print "This is invalid number".


```python
phoneBook = {
    'names': ['Amal', 'Mohammed', 'Khadijah', 'Abdullah', 'Rawan', 'Faisal', 'Layla'],
    'num': [1111111111, 2222222222, 3333333333, 4444444444, 5555555555, 6666666666, 7777777777]
}
ownerNum = input("Enter a phone number: ")
try:
    ownerNum = int(ownerNum)
    if not isinstance(ownerNum, int) or len(str(ownerNum)) != 10:
        print("This is invalid number")
    else:
        found = False
        for i in range(len(phoneBook['num'])):
            if phoneBook['num'][i] == ownerNum:
                print("The owner of the number:", phoneBook['names'][i])
                found = True
                break
        if not found:
            print("Sorry, the number is not found")
except ValueError:
    print("This is invalid number")
```

### Task 4
- Given the following list : [5, 4, 17, 19, 30, 2, 7, 10, 45]
    1. Use a lambda function to filter out only the odd numbers.
    2. Print the new list of odd numbers.

```python
# Given list
numbers = [5, 4, 17, 19, 30, 2, 7, 10, 45]

# Step 1: Use a lambda function to filter out only the odd numbers
odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))

# Step 2: Print the new list of odd numbers
print("List of odd numbers:", odd_numbers)
```
