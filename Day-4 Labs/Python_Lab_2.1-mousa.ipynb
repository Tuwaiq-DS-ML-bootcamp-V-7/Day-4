### Task 1
1. Write a function named is_prime that takes a number as a parameter and returns True if the number is prime and False otherwise. Print whether the number 29 is prime.

```python

def is_prime(number): #we use function named is_prime
    if number <= 1: 
            return False
    for i in range(2, number):
            if number % i == 0:
                return False
    return True

#number =int(input("Enter the number you want to check")) #input the number
number=29
is_prime(number)
```

### Task 3
Build a phone book program that receives the phone number, and returns the name of the owner.
You can follow the table below:

| Name     | Number     |
| -------- | ---------- |
| Amal     | 1111111111 |
| Mohammed | 2222222222 |
| Khadijah | 3333333333 |
| Abdullah | 4444444444 |
| Rawan    | 5555555555 |
| Faisal   | 6666666666 |
| Layla    | 7777777777 |

- If the number exists, print the owner. Otherwise, print "Sorry, the number is not found".
- If the number is less or more than 10 numbers, print "This is invalid number".
- If the number contains letters or symbols, print "This is invalid number".


```python
phone_book = {
    'Amal': 1111111111,
    'Mohammed': 2222222222,
    'Khadijah': 3333333333,
    'Abdullah': 4444444444,
    'Rawan': 5555555555,
    'Faisal': 6666666666,
    'Layla': 7777777777
}


def name_owner(number):
    
    check = str(number)#important  to check
    

    if len(check) < 10 or len(check) > 10 or not isinstance(number, int):
        return "This is invalid number"
    

    for name, phone in phone_book.items():
        if number == phone:
            return f'the Owner is : {name}'
    return "Sorry, the number is not found"


print(name_owner(6666666666))
```

### Task 4
- Given the following list : [5, 4, 17, 19, 30, 2, 7, 10, 45]
    1. Use a lambda function to filter out only the odd numbers.
    2. Print the new list of odd numbers.

```python
list_num = [5, 4, 17, 19, 30, 2, 7, 10, 45]
odd_numbers = list(filter(lambda x: x % 2 != 0, list_num))

print(odd_numbers)
```
